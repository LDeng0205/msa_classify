```python
### specify which msa sequences to train the model on
START_IDX = 40000
END_IDX = 80000
FILE_SUFFIX = (str(START_IDX)+'_'+ str(END_IDX)) if START_IDX != 0 else str(END_IDX)
```

```python
fasta_sequences = SeqIO.parse(open("PABP_YEAST/data/PABP_YEAST.a2m"),'fasta')
with open(f"PABP_YEAST/data/PABP_YEAST_{FILE_SUFFIX}.a2m", "w+") as out_file:
    cnt = 0
    seq = []
    for fasta in fasta_sequences:
        if cnt >= END_IDX:
            break
        if START_IDX <= cnt:
            seq.append(fasta)
        cnt += 1

    SeqIO.write(seq, out_file, "fasta")
```
